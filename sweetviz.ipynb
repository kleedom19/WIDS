````python
#using sweetviz for all kaggle data  
import os
from pathlib import Path
import numpy as np
import pandas as pd
import sweetviz as sv

# Sweetviz workaround for some NumPy versions
if not hasattr(np, "VisibleDeprecationWarning"):
    np.VisibleDeprecationWarning = DeprecationWarning

FILES = [
    "evac_zones_gis_evaczone.csv",
    "evac_zones_gis_evaczonechangelog.csv",
    "fire_perimeters_gis_fireperimeter.csv",
    "fire_perimeters_gis_fireperimeterchangelog.csv",
    "geo_events_externalgeoevent.csv",
    "geo_events_externalgeoeventchangelog.csv",
    "geo_events_geoevent.csv",
    "geo_events_geoeventchangelog.csv",
    "evac_zone_status_geo_event_map.csv",
]

OUT_DIR = Path("sweetviz_reports")
OUT_DIR.mkdir(parents=True, exist_ok=True)

generated = []

for f in FILES:
    csv_path = Path(f)
    if not csv_path.exists():
        print(f"SKIP (not found): {csv_path}")
        continue

    print(f"Reading: {csv_path}")
    df = pd.read_csv(csv_path, low_memory=False)

    base = csv_path.stem
    out_path = OUT_DIR / f"{base}_sweetviz.html"

    print(f"Analyzing -> {out_path}")
    report = sv.analyze(df)
    report.show_html(str(out_path), open_browser=False)

    generated.append((base, out_path))

# One-click index page
index_path = OUT_DIR / "index.html"
with open(index_path, "w", encoding="utf-8") as fh:
    fh.write("<!doctype html><html><head><meta charset='utf-8'>"
             "<title>Sweetviz Reports</title></head><body>")
    fh.write("<h1>Sweetviz Reports</h1><ul>")
    for base, path in generated:
        fh.write(f"<li><a href='{path.name}'>{base}</a></li>")
    fh.write("</ul></body></html>")

print(f"\nDone. Open: {index_path}")
````
